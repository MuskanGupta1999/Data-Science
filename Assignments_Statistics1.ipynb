# Calculate the mean, median, mode, and standard deviation for these columns.

```python
Avg_Price = df['Avg Price'].mean()
Total_Sales_Value = df['Total Sales Value'].mean()
Discount_Amount = df['Discount Amount'].mean() 
Net_Sales_Value = df['Net Sales Value'].mean()
Discount_Rate = df['Discount Rate (%)'].mean()
print(Avg_Price)
print(Total_Sales_Value)
print(Discount_Amount)
print(Net_Sales_Value)
print(Discount_Rate)
```

```python
Avg_Price = df['Avg Price'].median()
Total_Sales_Value = df['Total Sales Value'].median()
Discount_Amount = df['Discount Amount'].median() 
Net_Sales_Value = df['Net Sales Value'].median()
Discount_Rate = df['Discount Rate (%)'].median()
print(Avg_Price)
print(Total_Sales_Value)
print(Discount_Amount)
print(Net_Sales_Value)
print(Discount_Rate)
```

```python
Avg_Price = df['Avg Price'].mode()
Total_Sales_Value = df['Total Sales Value'].mode()
Discount_Amount = df['Discount Amount'].mode() 
Net_Sales_Value = df['Net Sales Value'].mode()
Discount_Rate = df['Discount Rate (%)'].mode()
print(Avg_Price)
print(Total_Sales_Value)
print(Discount_Amount)
print(Net_Sales_Value)
print(Discount_Rate)
```

```python
Avg_Price = df['Avg Price'].var()
Total_Sales_Value = df['Total Sales Value'].var()
Discount_Amount = df['Discount Amount'].var() 
Net_Sales_Value = df['Net Sales Value'].var()
Discount_Rate = df['Discount Rate (%)'].var()
print(Avg_Price)
print(Total_Sales_Value)
print(Discount_Amount)
print(Net_Sales_Value)
print(Discount_Rate)
```

```python
Avg_Price = df['Avg Price'].std()
Total_Sales_Value = df['Total Sales Value'].std()
Discount_Amount = df['Discount Amount'].std() 
Net_Sales_Value = df['Net Sales Value'].std()
Discount_Rate = df['Discount Rate (%)'].std()
print(Avg_Price)
print(Total_Sales_Value)
print(Discount_Amount)
print(Net_Sales_Value)
print(Discount_Rate)
```

```python
Avg_Price = df['Avg Price'].kurt()
Total_Sales_Value = df['Total Sales Value'].kurt()
Discount_Amount = df['Discount Amount'].kurt() 
Net_Sales_Value = df['Net Sales Value'].kurt()
Discount_Rate = df['Discount Rate (%)'].kurt()
print(Avg_Price)
print(Total_Sales_Value)
print(Discount_Amount)
print(Net_Sales_Value)
print(Discount_Rate)
```

# Bar Chart Analysis for Categorical Column:

```python
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

# Conversion of Categorical Data into Dummy Variables

```python
# 1. Need for Dummy Variables:
Categorical variables need to be converted into a numerical format for ML algorithms to process them effectively.
```

```python
# 2. Apply One-Hot Encoding:
df_dummies = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
print(df_dummies.head())
```
